##**Struktur Data Tuple**

Tuple tidak jauh beda dengan list yaitu sekumpulan data urutan yang berisi satu elemen atau lebih dari tipe data yang berbeda. Bedanya adalah terletak pada sifatnya yang immutable, artinya nilai atau elemen yang berada pada tuple tidak dapat dirubah. Tuple menggunakan tanda kurung biasa () dalam penulisannya dan setiap elemennya dipisahkan oleh koma.

```python
# contoh tuple kumpulan integer
tuple1 = (1, 5, 2, 3, 9)
print(tuple1)
```

```python
# tuple kumpulan berbagai tipe data
tuple2 = (1, 2.0, 'tiga')
print(tuple2)
```

```python
# tuple tanpa tanda kurung
tuple3 = 2, 4, 6, 8
tuple4 = 'a', 'b', 'c', 'd'

print(tuple3)
print(tuple4)
```

```python
# tuple kososng
tuple5 = ()
print(tuple5)
```

Untuk membuat tuple yang berisikan satu elemen, harus menuliskan tanda koma dibelakangnya. Jika tidak maka elemen tersebut tidak akan dianggap sebagai tuple, tetapi akan dianggap sebagai tipe data lain.

```python
# tuple berisikan satu elemen
tuple6 = (3,)
print(tuple6)
```

Perbedaan ketika tidak menuliskan koma

```python
tuple7 = (3,)   # ini tuple
tuple8 = (3)    # ini dianggap integer

print(tuple7, type(tuple7))
print(tuple8, type(tuple8))
```

###**Slicing Data pada Tuple**

Karena tuple mirip seperti list dan menyerupai array maka tuple juga mendukung proses slicing data, jadi kita dapat mengakses sebagaian atau lebih elemen yang ada pada tuple. Caranya sama persis dengan list.

```python
angka = (1, 8, 3, 4, 2, 5, 7, 6, 9, 10, 1)
print(angka)
```

```python
# Mengambil elemen indeks ke 0
print(angka[0])

# Mengambil elemen indeks ke 3
print(angka[3])

# Mengambil elemen indeks ke 3 dari belakang
print(angka[-3])

# Mengambil elemen indeks paling awal sapai indeks ke 5
print(angka[:6])

# Mengambil elemen indeks ke 6 sampai indeks terakhir
print(angka[6:])

# Mengambil elemen indeks ke 1 sampai indeks ke 8
print(angka[1:9])

# Mengambil elemen indeks ke 0 sampai ke 13 dengan langkah 2
print(angka[0:13:2])

# Mengambil keseluruhan data
print(angka[:])
```

###**Mengubah Elemen Data**

Karena tuple bersifat immutable (tidak dapat dirubah) maka nilai atau elemen didalam tuple tidak dapat dirubah. Artinya nilainya akan tetap. Jika kita memaksakan untuk merubah isi dalam tuple maka akan menghasilkan error.

```python
# merubah tuple, pasti error
ubah = ('merah', 'kuning', 'hijau', 'biru')
ubah[0] = 'putih'
```

Hal ini juga berlaku dalam menghapus elemen dalam tuple, secara otomatis akan menghasilkan error

```python
huruf = ('a', 'b', 'c', 'd')

# hapus indeks ke-1
del huruf[1]
```

Namun, jika yang dihapus itu keseluruhan elemen dalam tuple, maka hal tersebut tetap bisa dilakukan dengan memanfaatkan keyword del

```python
mapel = ('Matematika', 'Kimia', 'Fisika')
print(mapel)
```

```python
# hapus tuple mapel
del mapel
```

###**Beberapa Fungsi Bawaan Tuple**


*   len() : Menghitung jumlah elemen pada tuple
*   max() : Mencari nilai terbesar pada tuple
*   min() : Mencari nilai terkeil pada tuple




```python
bil_bulat = (2, 5, 4, 9, 1, 7)
print(bil_bulat)
```

```python
# Menghitung jumlah elemen tuple
len(bil_bulat)
```

```python
# Mencari nilai terbesar pada tuple
max(bil_bulat)
```

```python
# Mencari nilai terkecil pada tuple
min(bil_bulat)
```
