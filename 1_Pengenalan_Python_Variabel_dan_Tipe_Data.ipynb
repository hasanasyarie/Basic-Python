##**Variabel**

Sebelum memasuki tahap membuat sebuah code, alangkah baiknya kita mengenal sebuah variabel. Apa itu variabel? variabel adalah sebuah tempat atau wadah untuk menyimpan data atau nilai.

Contoh penulisan variabel:

```python
var = 'Hello World!'
```

Untuk memunculkan isi dari variabel cukup dengan menuliskan nama variabel tersebut

```python
var
```

Aturan penulisan variabel pada python: 

*   Variabel hanya boleh mengandung karakter alfabet, bilangan, dan underscore (a-z, A-Z, 0-9, _)
*   Nama variabel harus dimulai dengan huruf (a-z, A-Z) atau underscore
*   Nama variabel tidak boleh diawali oleh angka
*   Nama variabel tidak boleh mengandung spasi
*   Nama variabel tidak boleh mengandung reserverd keywords


Penulisan variabel yang benar:

```python
Nama = 'Guido van Rossum'
nama = 'Guido van Rossum'
_nama = 'Guido van Rossum'
nama1 = 'Guido van Rossum'
```

Penulisan variabel yang salah:

```python
1nama = 'Guido van Rossum'          #Diawali dengan angka
nama depan = 'Guido van Rossum'     #Mengandung spasi
nama% = 'Guido van Rossum'          #Mengandung simbol
in = 'Guido van Rossum'             #Mengandung reserverd keywords
```

Nama variabel pada python itu bersifat *case-sensitif*. Artinya huruf besar dan kecil dibedakan. Misalnya, **Nama** dan **nama** walaupun kalimatnya sama tapi itu dianggap variabel yang berbeda.

##**Tipe Data pada Python**

Terdapat beberapa tipe data pada python antara lain yaitu:
1. Integer (Bilangan Bulat)
2. Floating (Bilangan Desimal)
3. String (Karakter)
4. Boolean (True/False)
5. Blangan Kompleks

```python
var1 = 'Hello World!'          # string
var2 = 1234567890              # integer
var3 = 0.123456789             # float
var4 = True                    # boolean
var5 = 1j+1                    # complex
```

Memunculkan isi variabel dengan fungsi print():

```python
print(var1)
print(var2)
print(var3)
print(var4)
print(var5)
```

Mengecek tipe data:

```python
print(type(var1))
print(type(var2))
print(type(var3))
print(type(var4))
print(type(var5))
```

##**Konversi Tipe Data**

Konversi tipe data akan memudahkan kita untuk mengganti tipe data, misal kita ingin merubah tipe nilai dari suatu data dan mungkin melelahkan jika harus bolak balik mengganti nilai. Maka dari itu kita cukup mengkonversi data tersebut ke tipe data yang lain menggunakan sebuah fungsi.

Macam-macam fungsi untuk konversi tipe data:
1. int() = untuk mengkonversi ke tipe data integer
2. float() = untuk mengkonversi ke tipe data float
3. str() = untuk mengkonversi ke tipe data string
4. bool() = untuk mengkonversi ke tipe data boolean
5. long() = untuk mengubah ke tipe data integer panjang
6. chr() = untuk mengubah menjadi karakter
7. bin() = untuk mengubah menjadi bilangan biner
8. hex() untuk mengubah menjadi bilangan Heksadesimal
9. oct() untuk mengubah menjadi bilangan okta

```python
nilai = 15
print(nilai, 'Tipe data: ', type(nilai))
```

Konversi nilai yang bertipe integer menjadi float :

```python
konversi_float = float(nilai)
print(konversi_float, 'Tipe datanya ganti jadi :', type(konversi_float))
```

Konversi nilai yang bertipe integer menjadi string :

```python
konversi_str = str(nilai)
print(konversi_str, 'Tipe datanya ganti jadi :', type(konversi_str))
```

```python
a = oct(nilai)
a
```

```python
#mengkonversi nilai diatas
konv_float = float(nilai)
konv_string = str(nilai)
konv_boolean = bool(nilai)
konv_biner = bin(nilai)
konv_hexsa = hex(nilai)
konv_oct = oct(nilai)

#menampilkan output
print(konv_float, type(konv_float))
print(konv_string, type(konv_string))
print(konv_boolean, type(konv_boolean))
print(konv_biner, type(konv_biner))
print(konv_hexsa, type(konv_hexsa))
print(konv_oct, type(konv_oct))
```
