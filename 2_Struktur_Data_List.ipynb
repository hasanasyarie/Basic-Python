##**Struktur Data List**

List adalah sekumpulan data terurut dan bersifat mutabel (dapat berubah) yang didalamnya memiliki satu atau lebih nilai dari tipe data berbeda. Artinya dalam sebuah list diperbolehkan memiliki berbagai jenis tipe data yang berbeda-beda seperti string, integer, floating, dll. Tidak hanya itu, dalam suatu list juga dapat berisikan list didalam list. Struktur data list biasanya ditulis dengan kurung siku [] untuk menandainya dalam sebuah kode python.

```python
#Contoh list berisi tipe data berbeda
list1 = [1, 2, 3, 4.5, 'kalimat']
print(list1)
```

```python
#Contoh list didalam list
list2 = [1, 2, 3, [3, 2, 1], 'belajar']
print(list2)
```

```python
#Contoh list sekumpulan bilangan integer
list_int = [4, 3, 2, 1]
print(list_int)
```

```python
#Contoh list sekumpulan bilangan float
list_float = [0.5, 1.2, 3.4, 2.0]
print(list_float)
```

**Note:** Setiap elemen data pada list diawali dengan indeks yang dimulai dari 0. Contohnya seperti pada variabel list diatas jadi angka 1 itu indeks ke-0 kemudian angka 2 indeks ke-1 dan begitu seterusnya.

###**Slicing pada List**

Struktur data list mendukung isitilah slicing yang maksudnya adalah kita dapat menampilkan beberapa elemen data pada range tertentu dari elemen yang berada dalam list


*   x[0] yaitu mengambil elemen dengan indeks ke-0, artinya mengambil elemen pertama pada variabel list
*   x[6] yaitu mengambil elemen dengan indeks ke-6, jika kita hitung dari 0 maka artinya kita mengambil elemen ke-7 dari variabel list
*   x[-1] yaitu mengambil elemen paling belakang dari variabel list
*   x[:5] yaitu mengambil elemen dimulai dari elemen awal atau indeks ke-0 sampai ke elemen sebelum indeks ke-5 (dalam hal ini berarti mengambil elemen indeks ke-0 sampai elemen indeks ke-4)
*   x[2:6] yaitu mengambil elemen dimulai dari elemen pada indeks ke-1 sampai elemen indeks ke-5]
*   x[2:] yaitu mengambil elemen dimulai dari indeks ke-2 sampai elemen pada indeks terakhir
*   x[1:7:2] yaitu mengambil elemen pada indeks ke-1 sampai elemen pada indeks ke-6 dengan langkah 2 (artinya yang diambil hanya indeks ke 1, 3, dan 5)

Note: </br>
1. Indeks positif menandakan bahwa yang diambil adalah elemen dari kiri atau depan
2. Indeks negatif menandakan bahwa yang diambil adalah elemen dari kanan atau belakang
3. Indeks negatif tidak dimulai dari 0, tapi dari 1.
4. Titik dua sebelum angka menandakan bahwa yang diambil adalah elemen indeks dari kiri atau depan sampai elemen sebelum indeks yang ditetapkan
5. Titik dua setelah angka menandakan yang diambil adalah elemen yang dimulai dari indeks yang ditetapkan sampai dengan indeks terakhir 

**Contoh slicing pada list**

```python
list3 = [1, 8, 3, 4, 2, 5, 7, 6, 9, 10, 1]
print(list3)
```

```python
print(list3[0])
print(list3[3])
print(list3[-3])
print(list3[:6])
print(list3[6:])
print(list3[:-5])
print(list3[-5:])
print(list3[1:9])
print(list3[0:13:2])
print(list3[:])
```

Karena list itu bersifat mutabel maka setiap elemen didalam nya dapat kita rubah baik itu diganti, ditambah, ataupun dikurang.

###**Mengganti elemen/data**

```python
list_hewan = ['Kucing', 'Sapi', 'Kelinci', 'Ayam', 'Kambing']
print(list_hewan)

#Mengganti data pertama
list_hewan[0] = 'Tikus'
print(list_hewan)

#Mengganti data paling tengah
list_hewan[2] = 'Marmut'
print(list_hewan)

#Mengganti data terakhir
list_hewan[-1] = 'Bebek'
print(list_hewan)
```

###**Menambah elemen/data**

Ada beberapa fungsi yang bisa digunakan untuk menambah elemen dari sebuah list, antara lain seperti append(), extend(), dan insert().

####**Fungsi append()**

Fungsi append() berfungsi untuk menambahkan elemen dari belakang. Fungsi ini menerima satu parameter, yang mana parameter tersebut akan dimasukkan sebagai elemen baru pada list yang akan ditambahkan pada akhir elemen.

```python
#Menambahkan menggunakan fungsi append
list_angka = [11, 13, 15, 16, 17]
print(list_angka)

list_angka.append(18)
print(list_angka)
```

####**Fungsi insert()**

Berbeda dengan append() yang menambah elemen baru dari belakang, pada insert() kita dapat menambahkan elemen baru dimanapun dengan cara menuliskan indeksnya. Fungsi ini menerima 2 buah parameter, parameter pertama untuk menentukan indeks dan parameter kedua untuk mendefinisikan elemen yang akan ditambahkan ke dalam list.

```python
#Menambahkan dengan fungsi insert
list_buah = ['Apel', 'Jambu', 'Nanas', 'Melon']
print(list_buah)

list_buah.insert(3, 'Jeruk')
print(list_buah)
```

####**Fungsi extend()**

Fungsi extend() berfungsi untuk menambahkan beberapa elemen sekaligus pada list. Parameter yang dimasukan haruslah berupa nilai yang bersifat sequencial (berurut) seperti string atau list.

```python
# Menambahkan dengan fungsi extend()
x = [4, 3, 2]
y = [5, 6]

# Menambahkan list y ke dalam list x
x.extend(y)
print(x)

# Menambahkan huruf 'a' ke dalam list x
x.extend('a')
print(x)

# Menambahkan list yang berisi huruf 'a' dan 'b' ke dalam list x
x.extend(['b', 'c'])
print(x)

# Menambahkan angka 1 ke dalam list x
x.extend([1])
print(x)
```

###**Menghapus elemen/data**

Untuk menghapus elemen pada list dapat menggunakan keyword del atau menggunakan fungsi remove() dan fungsi pop().

####**Keyword del**

Keyword del menghapus elemen dalam list dengan cara memilih index elemen, konsepnya seperti silicing data jadi kita juga dapat menghapus beberapa elemen sekaligus.

```python
list_warna = ['Merah', 'Kuning', 'Hijau', 'Biru', 'Ungu', 'Coklat', 'Hitam']
print(list_warna)

# hapus index ke-0
del list_warna[0]
print(list_warna)

# hapus index ke 1-3
del list_warna[1:4]
print(list_warna)
```

####**Fungsi remove**

Fungsi remove() menghapus elemen dalam list dengan cara memilih elemen yang ingin kita hapus dengan menuliskannya didalam kurung. Berbeda dengan del, jika del menghapus dengan cara menuliskan indeks nya sedangkan remove() menghapus dengan cara spesifik menuliskan elemennya

```python
#Menggunakan fungsi remove()
list_kota = ['Jakarta', 'Bogor', 'Bekasi', 'Karawang']
print(list_kota)

list_kota.remove('Jakarta')
print(list_kota)
```

####**Fungsi pop**

Fungsi pop() sama seperti del yaitu menghapus elemen dengan menggunakan index. Perbedaannya adalah pada fungsi pop jika tidak menuliskan parameter indeks maka secara default akan menghapus elemen paling belakang.

```python
# Menggunakan metode pop()
list_huruf = ['a', 'b', 'c', 'd', 'e']
print(list_huruf)

list_huruf.pop(0)
print(list_huruf)
```

```python
# Tanpa menuliskan parameter indeks
list_huruf.pop()
print(list_huruf)
```

###**Mengurutkan dan membalikan urutan data**

Untuk mengurutkan elemen dari list kita dapat menggunakan fungsi sort(), sedangkan untuk membalikan urutan data dapat menggunakan fungsi reverse().

####**Fungsi sort()**

```python
# Mengurutkan data dengan fungsi sort()
urut = [3, 8, 1, 4, 0, 2, 7, 2]
print(urut)

urut.sort()
print(urut)
```

####**Fungsi reverse()**

```python
# Membalikan urutan dengan fungsi reverse()
urut.reverse()
print(urut)
```
